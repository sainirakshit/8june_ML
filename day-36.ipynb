```python
inert = []
for i in range(2,11):
    model_i = KMeans(n_clusters=i).fit(dummy)
    inert.append(model_i.inertia_)
```

```python

plt.figure(figsize=(18,6))
plt.plot(range(2,11), inert,marker='o')
plt.grid()
plt.xlabel('N_clusters')
plt.ylabel('Inertia')
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
cancer_sc = StandardScaler().fit_transform(cancer['data'])
```

```python
pd.DataFrame(cancer_sc).head()
```

```python
pd.DataFrame(cancer_sc).describe()
```

```python
inertia = []
for i in range(2,20):
    model_c = KMeans(n_clusters=i).fit(cancer_sc)
    inertia.append(model_c.inertia_)
plt.figure(figsize=(18,6))
plt.plot(range(2,20), inertia,marker='o')
plt.grid()
plt.xticks(range(2,20))
plt.xlabel('N_clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
inertia
```

```python
cancer['data'].shape
```

```python
cancer_sc.shape
```

```python
cancer_pca = PCA(n_components=2).fit_transform(cancer_sc)
```

```python
cancer_pca.shape
```

```python
plt.scatter(cancer_pca[:,0],cancer_pca[:,1],c=cancer['target'])
```

```python
sum(cancer['target'])
```

```python
inertia = []
for i in range(2,20):
    model_c = KMeans(n_clusters=i).fit(cancer_pca)
    inertia.append(model_c.inertia_)
plt.figure(figsize=(18,6))
plt.plot(range(2,20), inertia,marker='o')
plt.grid()
plt.xticks(range(2,20))
plt.xlabel('N_clusters')
plt.ylabel('Inertia')
plt.show()
```

```python

inertia
```

```python

kmeans_model = KMeans(n_clusters=2).fit(cancer_pca)
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.scatter(cancer_pca[:,0],cancer_pca[:,1],c=cancer['target'])
plt.title('Actual')
plt.subplot(1,2,2)
plt.scatter(cancer_pca[:,0],cancer_pca[:,1],c=kmeans_model.labels_)
plt.title('By KMeans')
plt.show()
```
